## 변수 연산

변수 연산은 우리가 이제 텍스트 + 텍스트 아니면 숫자 + 숫자 할때 사용됩니다!

어떻게 하냐고요?

아래와 같이 작동합니다!

```py
name = "Mireu/미르"
print("안녕하세요 저는 지나가던 개발자 " + name + " 입니다")
```
텍스트는 반드시 텍스트와 텍스트가있을때는 `+` 을 하게 되어있습니다! 왜냐면 텍스트 타입이라서 작동이 되근요!

숫자 넣으면 에러 떠요

이렇게요



~~작동되는게 더 이상하겠다~~

그래서 숫자을 넣을때는 택스트 사이에 `,`을 이용해서 실행을 시킵니다!

바로 이렇게요!

```py
age = 19
print("저의 나이는 지나가던 나이만큼" , age , "살 입니다~")
```
